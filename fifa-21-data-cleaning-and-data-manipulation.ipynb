## Drop Duplicates

```python
fifa_raw.drop_duplicates
fifa_raw.shape
```

After dropping, the number of row remain the same. There is no duplicates in the database.

## Replace unwanted line break in column 'Club'

```python
fifa_raw['Club'] = fifa_raw['Club'].replace('\n', '', regex=True)
fifa_raw['Club']
```

## Covert column 'Joined' and 'Loan Data End' into date time objects

```python
fifa_raw['Joined'] = pd.to_datetime(fifa_raw['Joined'])
fifa_raw['Day'] = fifa_raw['Joined'].dt.day
fifa_raw['Month'] = fifa_raw['Joined'].dt.month
fifa_raw['Year'] = fifa_raw['Joined'].dt.year
fifa_raw[['Joined','Day','Month','Year']]
```

We can do analyst of year or month that player join the club.

## Convert 'height' and 'weight' to numerical forms


First, we should understand the values in 'Height' and see how should they convert into numbers.

```python
height_counts = fifa_raw['Height'].value_counts()
print(height_counts)
```

There is two main issues in these values.
1. The values have units with them.
1. There are two different types of unit. We need to convert them into one.

```python
def convert_height(height:str):
    if "cm" in height:
        return int(height.replace('cm',''))
    elif "'" in height:
        feet_inch = re.split("\'", height)
        return int(round(float(feet_inch[0])* 30.48 + float(feet_inch[1].replace('"', ''))* 2.54))

fifa_raw['Height'] = fifa_raw['Height'].apply(convert_height)

height_counts = fifa_raw['Height'].value_counts()
print(height_counts)
```

```python
weight_counts = fifa_raw['Weight'].value_counts()
print(weight_counts)
```

The same as height the unit and type of units is still a problem. They have to convert into kg as unit.

```python
def convert_weight(weight: str):
    if "kg" in weight:
        return int(weight.replace('kg',''))
    elif "lb" in weight:
        lbs = int(weight.replace('lbs',''))
        return int(round(lbs*0.453592))

fifa_raw['Weight'] = fifa_raw['Weight'].apply(convert_weight)

weight_counts = fifa_raw['Weight'].value_counts()
print(weight_counts)
```

## Covert 'Value','Wage' and 'Release Clause; into numerical forms

These wages are in Euro and display using M or K. We need to convert them into numerical values.

```python
def convert_money(money: str):
    if '€' in money:
        money=money.replace("€","")
    if 'M' in money:
        return int(float(money.replace('M',''))*1000000)
    elif 'K' in money:
        return int(float(money.replace('K',''))*1000)

fifa_raw['Value'] = fifa_raw['Value'].apply(convert_money)
fifa_raw['Wage'] = fifa_raw['Wage'].apply(convert_money)
fifa_raw['Release Clause'] = fifa_raw['Release Clause'].apply(convert_money)
fifa_raw[['Value','Wage','Release Clause']]
```

## Remove stars in 'W/F','SM' and 'IR'

Weak foot, skill move and international reputation are stored as star level in data base. It is, however is not a good for numerical compare.

```python
fifa_raw['W/F'] = pd.to_numeric(fifa_raw['W/F'].replace('★','', regex=True))
fifa_raw['SM'] = pd.to_numeric(fifa_raw['SM'].replace('★','', regex=True))
fifa_raw['IR'] = pd.to_numeric(fifa_raw['IR'].replace('★','', regex=True))
fifa_raw[['W/F','SM','IR']]
```

## Covert 'Hits' into numerical forms

```python
fifa_raw['Hits']
```

```python
def covert_Hits(hits: str):
    hits = str(hits)
    if 'K' not in hits:
        return float(hits)
    else:
        return float(hits.replace('K',''))*1000
        
fifa_raw['Hits'] = fifa_raw['Hits'].apply(covert_Hits)
fifa_raw['Hits']
```

### Rename the column '↓OVA' to 'OVA'

```python
fifa_raw = fifa_raw.rename(columns={'↓OVA':'OVA'})
```

### Re-order of column and output

```python
fifa_cleaned = fifa_raw[['ID','Name','LongName','OVA','POT','Nationality','Club','Contract','Positions','Best Position','Joined','Day','Month','Year','Loan Date End','Value','Wage','Release Clause','Height','Weight','Preferred Foot','BOV','Attacking','Crossing','Finishing','Heading Accuracy','Short Passing','Volleys','Skill','Dribbling','Curve','FK Accuracy','Long Passing','Ball Control','Movement','Acceleration','Sprint Speed','Agility','Reactions','Balance','Power','Shot Power','Jumping','Stamina','Strength','Long Shots','Mentality','Aggression','Interceptions','Positioning','Vision','Penalties','Composure','Defending','Marking','Standing Tackle','Goalkeeping','GK Diving','GK Handling','GK Kicking','GK Positioning','GK Reflexes','Total Stats','Base Stats','W/F','SM','A/W','D/W','IR','PAC','SHO','PAS','DRI','DEF','PHY','Hits']]
fifa_cleaned
```
